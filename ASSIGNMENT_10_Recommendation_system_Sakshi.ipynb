# Book

Problem statement.

Build a recommender system by using cosine simillarties score.


```python
import pandas as pd
import numpy as np
```

```python
book_df = pd.read_csv(r'/content/book.csv',encoding="latin-1")
```

```python
book = book_df.iloc[:,1:]
```

```python
book
```

```python
book.columns = ['userID', 'title', 'bookRating']
```

```python
book
```

```python
#number of unique users in the dataset
len(book.userID.unique())
```

```python
len(book.title.unique())
```

```python
rating = pd.pivot_table(book, index='userID', 
                        values='bookRating', 
                        columns='title')
```

```python
rating
```

```python
rating.index=book.userID.unique()
```

```python
rating
```

```python
#Impute those NaNs with 0 values
rating.fillna(0, inplace=True)
```

```python
rating
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( rating.values,metric='cosine')
```

```python
user_sim 
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.userID.unique()
user_sim_df.columns = book.userID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
